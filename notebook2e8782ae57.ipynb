```python
#for test without submit
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

features = ["Pclass", "Sex", "FamilySize", "Is_child","Title"] 



X = pd.get_dummies(train_data[features])
y = train_data["Survived"]


X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=1)


model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X_train, y_train)


val_predictions = model.predict(X_val)


score = accuracy_score(y_val, val_predictions)
print(f"Senin Bilgisayarındaki Doğrulama Skoru: %{score * 100:.2f}")
```

```python
train_data["FamilySize"] = train_data["Parch"] + train_data["SibSp"] + 1
test_data["FamilySize"] = test_data["Parch"] + test_data["SibSp"] + 1

print(train_data[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index=False).mean())
```

```python
train_data['Title'] = train_data['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
test_data['Title'] = test_data['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

print(train_data['Title'].value_counts())
```

```python
rare_titles = ['Lady', 'Countess','Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona']

# Mapping (Haritalama) sözlüğü
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}

for dataset in [train_data, test_data]:
    # Nadirleri değiştir
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss') # Matmazel -> Miss
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')   # Madam -> Mrs
    dataset['Title'] = dataset['Title'].replace(rare_titles, 'Rare')
    
    # Kelimeleri sayıya çevir (Map)
    dataset['Title'] = dataset['Title'].map(title_mapping)
    
    dataset['Title'] = dataset['Title'].fillna(0)

print(train_data[['Name', 'Title']].head())
```

```python
features = ["Pclass", "Sex", "SibSp", "Parch", "Title"]


X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])
y = train_data["Survived"]

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)

predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)

print("Title özellikli final dosya 'submission.csv' olarak kaydedildi!")
```
